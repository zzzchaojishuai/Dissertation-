change number to text 

```python
def rearrange_confidenceLevel(data):
    switcher={
        1:'For',
        2:'For',
        3:'Neither',
        4:'Against',
        5:'Against',
    }
    return switcher.get(data,"NaN")


nhs['confidence_level'] = nhs['confidence_level'].apply(rearrange_confidenceLevel)
```

```python
def rearrange_region(data):
    switcher={
        1:'North_East',
        2:'North_West',
        3:'Yorkshire_and_The_Humber',
        4:'East_Midlands',
        5:'West_Midlands',
        6:'East_of_England',
        7:'London',
        8:'South_East',
        9:'South_West',
        10:'Wales',
        11:'Scotland',
    }
    return switcher.get(data,"NaN")


nhs['Region'] = nhs['Region'].apply(rearrange_region)
```

```python
def rearrange_gender(data):
    switcher={
        1:'Male',
        2:'Female',
    }
    return switcher.get(data,"NaN")


nhs['Gender'] = nhs['Gender'].apply(rearrange_gender)
```

```python
def rearrange_marriage(data):
    switcher={
        1:'Married',
        2:'Living_as_married',
        3:'Separated_after_marrying',
        4:'Widowed',
        5:'Not_married',

    }
    return switcher.get(data,"NaN")


nhs['Family'] = nhs['Family'].apply(rearrange_marriage)
```

```python
def rearrange_party(data):
    switcher={
        1:'Conservative',
        2:'Labour',
        3:'Liberal_Democrat',
        4:'Regional_Parties',
        5:'Regional_Parties',
        6:'Regional_Parties',
        7:'Other_Small_Parties',
        8:'Other_Small_Parties',
        9:'Other_Small_Parties',
        10:'Other_Small_Parties',
        11:'Other_Small_Parties',
        12:'None'
    }
    return switcher.get(data,"NaN")


nhs['Political'] = nhs['Political'].apply(rearrange_party)
```

```python
def rearrange_nation(data):
    switcher={
        1:'not British',
        2:'More (English/Scottish/Welsh) than British',
        3:'Equally (English/Scottish/Welsh) and British',
        4:'More British than (English/Scottish/Welsh)',
        5:'British not (English/Scottish/Welsh)',
        6:'East of England',
        7:'Other description ',
        8:'None of these',

    }
    return switcher.get(data,"NaN")


nhs['Nationity_1'] = nhs['Nationity_1'].apply(rearrange_nation)
```

```python
def rearrange_religion(data):
    switcher={
        1:'Church_of_England',
        2:'Roman_Catholic',
        3:'Other_Christian',
        4:'Non_Christian',
        5:'No_religion',
        6:'None',

    }
    return switcher.get(data,"NaN")


nhs['Religon'] = nhs['Religon'].apply(rearrange_religion)
```

```python
def rearrange_Nationity_2(data):
    switcher={
        1:'British',
        2:'English',
        3:'European',
        4:'Irish',
        5:'Northern Irish',
        6:'Scottish',
        7:'Ulster',
        8:'Welsh',
        9:'Other',
        10:'None',
    }
    return switcher.get(data,"NaN")


nhs['Nationity'] = nhs['Nationity'].apply(rearrange_Nationity_2)
```

```python
def rearrange_color(data):
    switcher={
        1:'BLACK: of African origin',
        2:'BLACK: of Caribbean origin',
        3:'BLACK: of other origin',
        4:'ASIAN: of Indian origin',
        5:'ASIAN: of Pakistani origin',
        6:'ASIAN: of Bangladeshi origin',
        7:'ASIAN: of Chinese origin',
        8:'ASIAN: of other origin',
        9:'WHITE: of any origin',
        10:'MIXED ORIGIN',
        11:'OTHER',
    }
    return switcher.get(data,"NaN")


nhs['Color'] = nhs['Color'].apply(rearrange_color)
```

```python
def rearrange_age(data):
    switcher={
        1:'18-25',
        2:'26-45',
        3:'26-45',
        4:'46-65',
        5:'46-65',
        6:'Above65',

    }
    return switcher.get(data,"NaN")


nhs['AgeGroup'] = nhs['AgeGroup'].apply(rearrange_age)
```
